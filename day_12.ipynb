```python
def twenty_generations(combos, initial_state, num_generations):
    num_pots = len(initial_state)
    generation = initial_state
    dots = -20
    for _ in range(num_generations):
        i = 0
        next_generation = ''
        while i < len(generation):
            if i == len(generation) - 2:
                seq = generation[i-2:]
                seq += '.'
            elif i == len(generation) - 1:
                seq = generation[i-2:]
                seq += '..'
            elif i == 0:
                seq = generation[-2:]
                last_chars = generation[:i+3]
                seq += last_chars
            elif i == 1:
                seq = generation[-1]
                last_chars = generation[:4]
                seq += last_chars
            else:
                seq = generation[i-2:i+3]
            next_generation += combos[seq]
            i += 1
        generation = '..' + next_generation + '..'
        dots -= 2
        yield sum([idx for idx, pound in enumerate(generation, dots) if pound == '#'])
```
